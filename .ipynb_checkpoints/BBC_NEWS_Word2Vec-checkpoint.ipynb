## Preprocessing of Data

```python
df.head()
```

```python
def clean_text(text):
    
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    #removing sentennce if its having only one word
    if len(text) > 2:
        return ' '.join(word for word in text.split() if word not in STOPWORDS)
    
df_clean = pd.DataFrame(df.text.apply(lambda x: clean_text(x)))
```

```python
df_clean.head()
```

```python
## Lemmatization using spacy
import en_core_web_sm
nlp = en_core_web_sm.load() ###Spacy version 3.0 link--> https://spacy.io/models

def lemmatizer(text):        
    sent = []
    doc = nlp(text)
    for word in doc:
        sent.append(word.lemma_)
    return " ".join(sent)
df_clean["text_lemmatize"] =  df_clean.apply(lambda x: lemmatizer(x['text']), axis=1)
```

```python
df_clean.head()
```

```python
df_clean['text_lemmatize_clean'] = df_clean['text_lemmatize'].str.replace('-PRON-', '') ##retaining the pro-nouns
```

```python
sentences = [row.split() for row in df_clean['text_lemmatize_clean']]
word_freq = defaultdict(int)
for sent in sentences:
    for i in sent:
        word_freq[i] += 1
len(word_freq)
```

```python
sorted(word_freq, key=word_freq.get, reverse=True)[:10]
```

## Word2Vec Model Building

```python
from gensim.models import Word2Vec
```

```python
## Different Parameters link--> https://radimrehurek.com/gensim/models/word2vec.html
# min_count: minimum number of occurrences of a word in the corpus to be included in the model.
# size: the dimensionality of the feature vectors
# workers: I know my system is having 6 cores, link https://support.microsoft.com/en-us/windows/find-out-how-many-cores-your-processor-has-3126ef99-0247-33b3-81fc-065e9fb0c35b

w2v_model = Word2Vec(min_count=200, window=5,size=100,workers=6)
```

```python
# this line of code to prepare the model vocabulary
w2v_model.build_vocab(sentences)
```

```python
# train word vectors
w2v_model.train(sentences, total_examples=w2v_model.corpus_count, epochs=w2v_model.iter)
```

```python
# As we do not plan to train the model any further, 
# we are calling init_sims(), which will make the model much more memory-efficient
w2v_model.init_sims(replace=True)
```

```python
# Explore the model
w2v_model.wv.most_similar(positive=['economy'])
```

```python
w2v_model.wv.most_similar(positive=['president'])
```

```python
# how similar are these two words to each other 
w2v_model.wv.similarity('company', 'business')
```

```python
def tsne_plot(model):
    "Create TSNE model and plot it"
    labels = []
    tokens = []

    for word in model.wv.vocab:
        tokens.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(18, 18)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```

```python
# Use t-SNE to represent high-dimensional data 
# and the underlying relationships between vectors in a lower-dimensional space.
tsne_plot(w2v_model)
```
